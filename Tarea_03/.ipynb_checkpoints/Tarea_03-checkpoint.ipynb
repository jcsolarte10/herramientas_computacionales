```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
def extraer_datos(tipo, num):
    if(tipo == "electrico"):
        datos = os.path.join( '.',os.path.join('data_'+ tipo,'VBS__'+str(num)+ '.VBO'))
        data_pd = pd.read_csv(datos, sep=' ',header=None , skiprows=69)

    if(tipo == "combustion"):
        datos = os.path.join( '.',os.path.join('data_'+ tipo,'VBOX'+str(num)+ '.VBO'))
        data_pd = pd.read_csv(datos, sep=' ',header=None , skiprows=35)        
    
    return data_pd
```
