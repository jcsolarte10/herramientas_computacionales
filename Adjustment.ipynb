```python
import numpy as np
from scipy.optimize  import curve_fit
import matplotlib.pyplot as plt
```

```python
def poly1(x,a,b):
    fun = a*x + b
    return fun


def poly2(x,a,b,c):
    fun = a*x**2+ b*x + c
    return fun

def pow_law(x,a,b):
    fun = a*x**b
    return fun
```

```python
gamma_dot = np.array([50, 70, 90,110,130])
tau = np.array([6.01, 7.48, 8.59, 9.19, 10.21])
```

```python
n,o = curve_fit(poly1,gamma_dot,tau)
error = np.sqrt(sum((poly1(gamma_dot,*n)-tau)**2)/gamma_dot.size)
print(error)
sam=np.linspace(gamma_dot[0],gamma_dot[-1],100)
plt.plot(sam,poly1(sam,*n))
plt.plot(gamma_dot,tau,'*k')
plt.show()
```

```python
n,o = curve_fit(poly2,gamma_dot,tau)
error = np.sqrt(sum((poly2(gamma_dot,*n)-tau)**2)/gamma_dot.size)
print(error)
sam=np.linspace(gamma_dot[0],gamma_dot[-1],100)
plt.plot(sam,poly2(sam,*n))
plt.plot(gamma_dot,tau,'*k')
plt.show()
```

```python
n,o = curve_fit(pow_law,gamma_dot,tau)
error = np.sqrt(sum((pow_law(gamma_dot,*n)-tau)**2)/gamma_dot.size)
print(error)
sam=np.linspace(gamma_dot[0],gamma_dot[-1],100)
plt.plot(sam,pow_law(sam,*n))
plt.plot(gamma_dot,tau,'*k')
plt.show()
```
