```python
%pylab inline
import scipy
from scipy import misc
from scipy import ndimage
from skimage import feature



from skimage.transform import (hough_line, hough_line_peaks,
                               probabilistic_hough_line)
```

```python
face = misc.face()
```

```python
plt.imshow(face)
plt.imsave('ej.png',face)
```

```python
crop_face = face[50:500,400:900, :]
plt.imshow(crop_face)
plt.show()
```

```python
ej = ndimage.imread('./ej.png', mode = 'F')
ej.dtype
```

```python
blurred_1 = ndimage.gaussian_filter(face, sigma=(1, 1, 0))
blurred_2 = ndimage.uniform_filter(face, size=(1, 1, 0))

subplot(1,2,1)
plt.imshow(blurred_1)
plt.axis('off')
subplot(1,2,2)
plt.imshow(blurred_2)
plt.axis('off')

plt.show()
```

```python
im = np.zeros((256, 256))
im[64:-64, 64:-64] = 100


im = ndimage.rotate(im, 15, mode='constant')
im = ndimage.gaussian_filter(im, 3)

sx = ndimage.sobel(ej, axis=0, mode='constant')
sy = ndimage.sobel(ej, axis=1, mode='constant')
sob = np.hypot(sx, sy)


plt.imshow(sob)
plt.show()
im.dtype
```
