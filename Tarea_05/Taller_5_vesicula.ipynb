```python
%pylab inline
import scipy
from scipy import misc
from scipy import ndimage
from skimage import feature
from skimage.draw import circle_perimeter


from skimage.transform import (hough_circle,hough_circle_peaks)
```

```python
ves = ndimage.imread('./ves_full.jpg', mode = 'F')
plt.imshow(ves, cmap = cm.gray)
plt.show()
```

```python
ves2 = ndimage.gaussian_filter(ves,2)
edges = feature.canny(ves2, sigma = 13)
plt.imshow(edges, cmap = cm.gray)
plt.show()
```

```python
radius = np.arange(150,400,1)
res = hough_circle(edges,radius)
acum,cx,cy,radii = hough_circle_peaks(res,radius,total_num_peaks=1)
```

```python
im = np.zeros((shape(ves)[0], shape(ves)[1]))
vx,vy = circle_perimeter(cx[0],cy[0],radii[0])
im[vx,vy] =100
plt.imshow(ves, cmap = cm.gray)
plt.scatter(vx,vy,color = "r", s = 0.5)
plt.scatter(cx[0],cy[0],color = "r", s = 2)
plt.axis('off')
plt.savefig("results.jpg")
plt.show()
```

Se encontro que:

El centroide de la vesícula se ubica en: (278 , 280)

El diametro de la vesícula es: 476 px
